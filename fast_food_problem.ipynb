```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
%matplotlib inline
```

```python
# These are pricess of fries and burger
# These will not be known to the NN, it will need to figure it out
w = [15, 25] 
```

```python
# Generate some training data
x = []
y = []

for i in range(100):
    x1 = np.random.randint(0, 5)
    x2 = np.random.randint(0, 5)
    x.append([x1, x2])
    y.append(x1 * w[0] + x2 * w[1])

x = np.array(x)
y = np.array(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
x[2], y[2]
```

```python
def error(prediction, ground_truth):
    return np.sum(np.square(prediction - ground_truth)) / 2.0
```

```python
pred = np.random.rand(y.shape[0]) * 100
```

```python
print(error(pred, y))
```

```python
wa = np.arange(0, 30, 1)
wb = np.arange(0, 30, 1)
X, Y = np.meshgrid(wa, wb)
er = np.zeros(X.shape)

for w1 in wa:
    for w2 in wb:
        w_forgraph = [w1, w2]
        er[w1][w2] = error(np.array(w_forgraph).dot(x.T), y)
        
for w1 in wb:
    for w2 in wa:
        w_forgraph = [w1, w2]
        er[w1][w2] = error(np.array(w_forgraph).dot(x.T), y)
```

```python
er[0][1]
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')

ax.plot_surface(wa, wb, er, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.show()
```

```python


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)


# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
