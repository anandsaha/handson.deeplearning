```python
from tensorflow.examples.tutorials.mnist import input_data
```

### MNIST with ANN's

Placeholders and Variables

```python
img_len = 784
num_digits = 10
K = 200
L = 100
M = 60
N = 30
```

```python
x = tf.placeholder(tf.float32, [None, img_len])
y_actual = tf.placeholder(tf.float32, [None, num_digits])
lr = tf.placeholder(tf.float32)
pkeep = tf.placeholder(tf.float32)

w1 = tf.Variable(initial_value=tf.truncated_normal([img_len, K], stddev=0.1), dtype=tf.float32)
b1 = tf.Variable(tf.ones([1, K]), dtype=tf.float32)

w2 = tf.Variable(initial_value=tf.truncated_normal([K, L], stddev=0.1), dtype=tf.float32)
b2 = tf.Variable(tf.ones([1, L]), dtype=tf.float32)

w3 = tf.Variable(initial_value=tf.truncated_normal([L, M], stddev=0.1), dtype=tf.float32)
b3 = tf.Variable(tf.ones([1, M]), dtype=tf.float32)

w4 = tf.Variable(initial_value=tf.truncated_normal([M, N], stddev=0.1), dtype=tf.float32)
b4 = tf.Variable(tf.ones([1, N]), dtype=tf.float32)

# SoftMax
w5 = tf.Variable(initial_value=tf.truncated_normal([N, num_digits], stddev=0.1), dtype=tf.float32)
b5 = tf.Variable(tf.zeros([1, num_digits]), dtype=tf.float32)
```

Model

```python
y1 = tf.nn.relu(tf.matmul(x, w1) + b1)
y1d = tf.nn.dropout(y1, pkeep)

y2 = tf.nn.relu(tf.matmul(y1, w2) + b2)
y2d = tf.nn.dropout(y2, pkeep)

y3 = tf.nn.relu(tf.matmul(y2, w3) + b3)
y3d = tf.nn.dropout(y3, pkeep)

y4 = tf.nn.relu(tf.matmul(y3, w4) + b4)
y4d = tf.nn.dropout(y4, pkeep)

logits = tf.matmul(y4, w5) + b5

y_predict = tf.nn.softmax(logits)
```

Loss and optimizer

```python
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=y_actual)
cross_entropy = tf.reduce_mean(cross_entropy) * 100
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate=lr)
trainner = optimizer.minimize(cross_entropy)
```

Accuracy

```python
corrects = tf.equal(tf.argmax(y_actual, 1), tf.argmax(y_predict, 1))
accuracy = tf.reduce_mean(tf.cast(corrects, tf.float32))
```

Training

```python
num_epochs = 100
batch_size = 100
```

```python
init = tf.global_variables_initializer()
```

```python
x_test = mnist.test.images.reshape((-1, img_len))
y_test = mnist.test.labels
```

```python
train_accuracy = list()
train_loss = list()

test_accuracy = list()
test_loss = list()
```

```python
with tf.Session() as sess:
    sess.run(init)
    for epoch in range(num_epochs):
        
        max_learn_rate = 0.03
        min_learn_rate = 0.001
        decay_speed = 2000
        learning_rate = min_learn_rate + (max_learn_rate - min_learn_rate) * math.exp(-epoch/decay_speed)
        learning_rate = 0.001
        
        for batch_num in range(mnist.train.num_examples // batch_size):
            x_batch, y_batch = mnist.train.next_batch(batch_size)
            x_batch = x_batch.reshape((-1, img_len))
            
            train_data = {x: x_batch, y_actual: y_batch, pkeep: 0.60, lr: learning_rate}
            sess.run(trainner, feed_dict=train_data)
        
        a_train, c_train = sess.run([accuracy, cross_entropy], feed_dict={x: x_batch, y_actual: y_batch})
        a_test, c_test = sess.run([accuracy, cross_entropy], feed_dict={x: x_test, y_actual:y_test})
        
        train_accuracy.append(a_train)
        train_loss.append(c_train)
        
        test_accuracy.append(a_test)
        test_loss.append(c_test)
        
        print('Epoch', epoch)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(15, 10))
plt.plot(train_accuracy)
plt.plot(test_accuracy)
plt.show()

fig = plt.figure(figsize=(15, 10))
plt.plot(train_loss)
plt.plot(test_loss)
plt.show()
```

### MNIST with CNN's

```python
x = tf.placeholder(tf.float32, [None, 28, 28, 1])
y_actual = tf.placeholder(tf.float32, [None, 10])
lr = tf.placeholder(tf.float32)
pkeep = tf.placeholder(tf.float32)
```

```python
C1 = 4 # Conv Layer 1
C2 = 8 # Conv Layer 2
C3 = 12 # Conv Layer 3
C4 = 200 # Fully connected
num_classes = 10
```

```python
w1 = tf.Variable(tf.truncated_normal([5, 5, 1, C1], stddev=0.1))
b1 = tf.Variable(tf.ones([C1])/10)

w2 = tf.Variable(tf.truncated_normal([5, 5, C1, C2], stddev=0.1))
b2 = tf.Variable(tf.ones([C2])/10)

w3 = tf.Variable(tf.truncated_normal([4, 4, C2, C3], stddev=0.1))
b3 = tf.Variable(tf.ones([C3])/10)


w4 = tf.Variable(tf.truncated_normal([7 * 7 * C3, C4], stddev=0.1))
b4 = tf.Variable(tf.ones([C4])/10)

w5 = tf.Variable(tf.truncated_normal([C4, num_classes], stddev=0.1))
b5 = tf.Variable(tf.ones([num_classes])/10)
```

```python
# The model

stride = 1
y1 = tf.nn.relu(tf.nn.conv2d(x, w1, strides=[1, stride, stride, 1], padding='SAME') + b1)

stride = 2
y2 = tf.nn.relu(tf.nn.conv2d(y1, w2, strides=[1, stride, stride, 1], padding='SAME') + b2)

stride = 2
y3 = tf.nn.relu(tf.nn.conv2d(y2, w3, strides=[1, stride, stride, 1], padding='SAME') + b3)

yy = tf.reshape(y3, [-1, 7 * 7 * C3])

y4 = tf.nn.relu(tf.matmul(yy, w4) + b4)

yy4 = tf.nn.dropout(y4, pkeep)

logits = tf.matmul(yy4, w5) + b5
y_predict = tf.nn.softmax(logits)

```

```python
# The loss
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=y_actual)
cross_entropy = tf.reduce_mean(cross_entropy) * 100
```

```python
correct_prediction = tf.equal(tf.argmax(y_actual, 1), tf.argmax(y_predict, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
# Optimizer
trainner = tf.train.AdamOptimizer(lr).minimize(cross_entropy)
```

```python
init = tf.global_variables_initializer()
```

```python
num_epochs = 100
batch_size = 100
```

```python
x_test = mnist.test.images
y_test = mnist.test.labels
```

```python
train_accuracy = list()
train_loss = list()

test_accuracy = list()
test_loss = list()
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    for epoch in range(num_epochs):
        for batch_num in range(mnist.train.num_examples // batch_size):
            x_batch, y_batch = mnist.train.next_batch(batch_size)
            sess.run(trainner, feed_dict={x: x_batch, y_actual: y_batch, lr:0.001, pkeep:0.75})
        a, c = sess.run([accuracy, cross_entropy], feed_dict={x: x_batch, y_actual: y_batch, pkeep:1.0})
        train_accuracy.append(a)
        train_loss.append(c)
        
        a, c = sess.run([accuracy, cross_entropy], feed_dict={x: x_test, y_actual: y_test, pkeep:1.0})
        test_accuracy.append(a)
        test_loss.append(c)
        
        print('Epoch', epoch)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(15, 10))
plt.plot(train_accuracy)
plt.plot(test_accuracy)
plt.show()

fig = plt.figure(figsize=(15, 10))
plt.plot(train_loss)
plt.plot(test_loss)
plt.show()
```
