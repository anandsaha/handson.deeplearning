```python
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
from tensorflow.examples.tutorials.mnist import input_data
data = input_data.read_data_sets("/home/anand/.tensorflow/data/MNIST/", one_hot=True)
```

```python
# Set dimensions

img_size = 28
img_size_flat = img_size * img_size
img_shape = (img_size, img_size)
num_classes = 10
```

```python
data.test.cls = np.array([label.argmax() for label in data.test.labels])
```

```python
def plot_image(nine_images, class_true, class_pred = None):
    fig, axes = plt.subplots(3, 3)
    fig.subplots_adjust(hspace=0.3, wspace=0.3)
    
    for i, ax in enumerate(axes.flat):
        ax.imshow(nine_images[i].reshape(img_size, img_size), cmap='binary')
        if class_pred is None:
            xlabel = 'True: {0}'.format(class_true[i])
        else:
            xlabel = 'True: {0}, Pred: {1}'.format(class_true[i], class_pred[i])
            
        ax.set_xlabel(xlabel)
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
plot_image(data.test.images[0:9, :], data.test.labels[0:9, :])
```

Placeholders

```python
x = tf.placeholder(tf.float32, [None, img_size_flat])
y_true = tf.placeholder(tf.int64, [None, num_classes])
y_true_class = tf.placeholder(tf.int64, [None])
```

Variables

```python
weights = tf.Variable(tf.zeros([img_size_flat, num_classes]))
biases = tf.Variable(tf.zeros([num_classes]))
```

Model

```python
logits = tf.matmul(x, weights) + biases
```

```python
y_pred = tf.nn.softmax(logits)
```

```python
y_pred_class = tf.argmax(y_pred, axis=1)
```

Cost

```python
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=y_true)
cost = tf.reduce_mean(cross_entropy)
```

Optimization

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.5).minimize(cost)
```

Performance measures

```python
correct_prediction = tf.equal(y_pred_class, y_true_class)
```

```python
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

Tensor execution stuff

```python
session = tf.Session()
```

```python
session.run(tf.global_variables_initializer())
```

```python
batch_size = 128
```

```python
def optimize(epochs):
    for i in range(epochs):
        x_batch, y_true_batch = data.train.next_batch(batch_size)
        session.run(optimizer, feed_dict={x:x_batch, y_true:y_true_batch})
```

```python
def print_accuracy():
    acc = session.run(accuracy, feed_dict={x: data.test.images, y_true: data.test.labels, 
                                           y_true_class: data.test.cls})
    
    print("Accuracy on test set: {0:.1%}".format(acc))
```

```python
def plot_weights():
    w = session.run(weights)
    
    w_max = np.max(w)
    w_min = np.min(w)
    
    fig, axes = plt.subplots(3, 4)
    fig.subplots_adjust(hspace=0.3, wspace=0.3)
    
    for i, ax in enumerate(axes.flat):
        if i < 10:
            image = w[:, i].reshape(img_shape)
            ax.set_xlabel("Weight: {0}".format(i))
            ax.imshow(image, vmin=w_min, vmax=w_max, cmap='jet')
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
def print_confusion_matrix():
    # Get the true classifications for the test-set.
    cls_true = data.test.cls
    
    feed_dict_test = {x: data.test.images,
                  y_true: data.test.labels,
                  y_true_class: data.test.cls}
    
    # Get the predicted classifications for the test-set.
    cls_pred = session.run(y_pred_class, feed_dict=feed_dict_test)

    # Get the confusion matrix using sklearn.
    cm = confusion_matrix(y_true=cls_true,
                          y_pred=cls_pred)
    
    # Print the confusion matrix as text.
    print(cm)
    
    cm = cm / np.sum(cm, axis=1)
    np.fill_diagonal(cm, 0)

    # Plot the confusion matrix as an image.
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)

    # Make various adjustments to the plot.
    plt.tight_layout()
    plt.colorbar()
    tick_marks = np.arange(num_classes)
    plt.xticks(tick_marks, range(num_classes))
    plt.yticks(tick_marks, range(num_classes))
    plt.xlabel('Predicted')
    plt.ylabel('True')
```

```python
optimize(1)
print_accuracy()
plot_weights()
```

```python
optimize(99)
print_accuracy()
plot_weights()
```

```python
optimize(900)
print_accuracy()
plot_weights()
```

```python
optimize(1000)
print_accuracy()
plot_weights()
```

```python
print_confusion_matrix()
```
