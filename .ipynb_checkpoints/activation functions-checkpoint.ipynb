### Sigmoid

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.array([i for i in range(-20, 21)])
```

```python
def sigmoid(i):
    return 1 / (1 + np.exp(-i))
```

```python
y = sigmoid(x)
```

```python
plt.plot(x, y)
plt.show()
```

### Tanh

```python
y = np.tanh(x)
```

```python
plt.plot(x, y)
plt.show()
```

### ReLU

```python
def relu(x):
    return np.maximum(0, x)
```

```python
y = relu(x)
```

```python
plt.plot(x, y)
plt.show()
```

### Softmax

```python
# Let's simulate an output where the second last item has the max score
x = [2, 3, 1, 4, 1, 3, 2, 6, 12, 8]
```

```python
def softmax(x):
    denominator = 0
    for i in x:
        denominator += np.exp(i)
    return np.exp(x) / denominator
```

```python
y = softmax(x)
```

```python
np.sum(y)
```

```python
plt.bar([i for i in range(len(x))], y)
plt.show()
```
